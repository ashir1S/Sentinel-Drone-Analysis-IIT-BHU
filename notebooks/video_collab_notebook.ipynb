###3: LOAD MODEL & DEFINE LOGIC

```python
from ultralytics import YOLO
import cv2
import numpy as np

# 1. LOAD YOUR TRAINED MODEL
print(f"🔄 Loading Neural Network...")
model = YOLO(MODEL_PATH)
print("✅ Model Active.")

# 2. DEFINE HUD SETTINGS
# Based on your diagnostic check:
# 0:camo, 1:weapon, 2:tank, 3:truck, 4:vehicle, 5:soldier, 6:artillery, 7:aircraft, 8:warship
LABELS = {
    0: 'CAMO-SOLDIER',
    1: 'WEAPON',
    2: 'TANK',
    3: 'TRUCK',
    4: 'VEHICLE',
    5: 'SOLDIER',
    6: 'ARTILLERY',
    7: 'AIRCRAFT',
    8: 'WARSHIP'
}

# HUD Colors (BGR format)
COLOR_RED = (0, 0, 255)    # Threat
COLOR_CYAN = (255, 255, 0) # Info

def get_grid_ref(x, y, w, h, cls_id):
    """Calculates the MGRS Grid Reference."""
    e = int((x / w) * 10000)
    n = int((y / h) * 10000)
    if cls_id == 8: return f"NAV-SEC {e:04d}-{n:04d}"
    return f"34S WC {e:04d} {n:04d}"
```

###4: VIDEO PROCESSOR

```python
import time

def run_sentinel_processing():
    # Define Output File
    output_filename = os.path.join(OUTPUT_DIR, "SENTINEL_HUD_33sec.mp4")

    print(f"\n▶ STARTING MISSION: 33sec.mp4")

    # Open Video
    cap = cv2.VideoCapture(VIDEO_PATH)
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Set up Writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_filename, fourcc, fps, (width, height))

    frame_count = 0
    start_time = time.time()

    while True:
        ret, frame = cap.read()
        if not ret: break

        # --- 1. AI DETECTION ---
        # conf=0.25 is a good balance for standard detection
        results = model(frame, verbose=False, conf=0.25)

        # --- 2. DRAW HUD ---
        for r in results:
            for box in r.boxes:
                # Extract Data
                cls_id = int(box.cls[0])
                x, y, x2, y2 = box.xywh[0].tolist()

                # Coordinates for drawing
                x1, y1 = int(x - x2/2), int(y - y2/2)
                x2_corn, y2_corn = int(x + x2/2), int(y + y2/2)

                # Get Label
                label_text = LABELS.get(cls_id, 'UNKNOWN')
                grid_ref = get_grid_ref(x, y, width, height, cls_id)

                # GRAPHICS: Corners
                l = int(min(x2, y2) / 4)
                thick = 2
                cv2.line(frame, (x1, y1), (x1+l, y1), COLOR_RED, thick)
                cv2.line(frame, (x1, y1), (x1, y1+l), COLOR_RED, thick)
                cv2.line(frame, (x2_corn, y2_corn), (x2_corn-l, y2_corn), COLOR_RED, thick)
                cv2.line(frame, (x2_corn, y2_corn), (x2_corn, y2_corn-l), COLOR_RED, thick)

                # GRAPHICS: Header (ID)
                cv2.rectangle(frame, (x1, y1-20), (x1+130, y1-2), COLOR_RED, -1)
                cv2.putText(frame, label_text, (x1+5, y1-6),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2)

                # GRAPHICS: Footer (Coordinates)
                text_size = cv2.getTextSize(grid_ref, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)[0]
                cv2.rectangle(frame, (x1, y2_corn + 2), (x1 + text_size[0] + 10, y2_corn + 20), (0,0,0), -1)
                cv2.putText(frame, grid_ref, (x1 + 5, y2_corn + 15),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

        # Save Frame
        out.write(frame)

        # Progress Bar (Updates every 30 frames)
        frame_count += 1
        if frame_count % 30 == 0:
            elapsed = time.time() - start_time
            print(f"   Processing: Frame {frame_count}/{total_frames} ({frame_count/elapsed:.1f} FPS)", end='\r')

    # Cleanup
    cap.release()
    out.release()
    print(f"\n\n✅ MISSION COMPLETE.")
    print(f"📂 Output saved to: {output_filename}")
```

###5: EXECUTE MISSION

```python
# Run the function defined above
run_sentinel_processing()
```
